# Nairobi Data

```python
# Retrieving weather data for Nairobi
from openmeteo_py import Hourly,Daily,Options,OWmanager

# Latitude, Longitude for Nairobi,Kenya

latitude = 1.2921
longitude = 36.8219

hourly = Hourly()
daily = Daily()
options = Options(latitude,longitude)

mgr = OWmanager(options,
    hourly.all(),
    daily.all())


# Download data
Nairobi = mgr.get_data()

print(Nairobi)
```

```python
#Creating pandas dataframe for Nairobi data
Nairobi_df = pd.DataFrame.from_dict(Nairobi) 
Nairobi_df=Nairobi_df.transpose()
#Selecting the required rows
Nairobi_df=Nairobi_df[['time','apparent_temperature', 'precipitation_sum', 
                             'precipitation_hours', 'windspeed_10m', 'windspeed_10m_max']]
#Insert New column 'City Name'
Nairobi_df['City_name'] = 'Nairobi'
#Display dataframe
Nairobi_df
```

```python
# Checking for null values 
Nairobi_df.isnull().sum()
```

```python
# Fill the NULL values with zeros
Nairobi_df = Nairobi_df.fillna(0)
Nairobi_df
```

```python
# Merge the Nairobi and Kisumu dataframes
frames = [Nairobi_df, Kisumu_df]
  
Nairobi_kisumu_df = pd.concat(frames)
Nairobi_kisumu_df=Nairobi_kisumu_df.reset_index()
Nairobi_kisumu_df=Nairobi_kisumu_df.drop(['index'], axis=1)
display(Nairobi_kisumu_df)
```

```python
# Merge the Nakuru and Mombasa dataframes
frames = [Nakuru_df, Mombasa_df]
  
Nakuru_Mombasa_df = pd.concat(frames)
Nakuru_Mombasa_df=Nakuru_Mombasa_df.reset_index()
Nakuru_Mombasa_df=Nakuru_Mombasa_df.drop(['index'], axis=1)
display(Nakuru_Mombasa_df)
```

```python
# Merge all the dataframes into one
frames = [Nairobi_kisumu_df, Nakuru_Mombasa_df]
  
Weather_data = pd.concat(frames)
Weather_data=Weather_data.reset_index()
Weather_data=Weather_data.drop(['index'], axis=1)
display(Weather_data)
```
